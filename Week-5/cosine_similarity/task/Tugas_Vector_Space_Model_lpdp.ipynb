```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset dari slide Stanford
corpus = [
    "Dari delapan orang tadi (yang dikenakan sanksi), empat orang sudah lunas bayar kas ke negara langsung." , # Doc 0
    "Namun, ketika sebagian penerima memilih jalur karier yang dinilai tidak sejalan dengan ketentuan pengabdian, publik mempertanyakan konsistensi kebijakan, mekanisme sanksi, serta keadilan bagi alumni lain yang patuh terhadap aturan.",                           # Doc 1
    "Selain delapan alumni yang telah dijatuhi sanksi, LPDP juga tengah memeriksa 36 alumni lain yang diduga melakukan pelanggaran kewajiban pengabdian."    # Doc 2
]

# 2. Representasi Vektor (Hanya fokus pada kata 'hate' dan 'darkness')
vectorizer = CountVectorizer(vocabulary=['alumni', 'sanksi'])
X = vectorizer.fit_transform(corpus)

# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['alumni', 'sanksi'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

Analisa Cosine Similarity
Doc 0: Jadi patokan utama karena skornya sempurna, yaitu 1
Doc 1: Skornya 0.70 karena ada satu kata yang sama dengan Doc 0, yaitu sanksi, akan tetapi arahnya mulai bergeser karena muncul kata alumni
Doc 2: Skornya paling renda karena jumlah kata alumninya lebih banyak, sehingga topiknya dianggap semakin jauh dari Doc 0 yang murni bahas sanksi

```python
documents = [
    "Dari delapan orang tadi (yang dikenakan sanksi), empat orang sudah lunas bayar kas ke negara langsung." , # Doc 0
    "Namun, ketika sebagian penerima memilih jalur karier yang dinilai tidak sejalan dengan ketentuan pengabdian, publik mempertanyakan konsistensi kebijakan, mekanisme sanksi, serta keadilan bagi alumni lain yang patuh terhadap aturan.",                           # Doc 1
    "Selain delapan alumni yang telah dijatuhi sanksi, LPDP juga tengah memeriksa 36 alumni lain yang diduga melakukan pelanggaran kewajiban pengabdian."    # Doc 2
]
```

```python
import pandas as pd
from collections import Counter
tf = pd.DataFrame(
[Counter(doc.split()) for doc in documents],
).fillna(0)
tf
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```

Hasil:
- Doc 0 ke Doc 1 (1.0): Hasil ini muncul apabila menggunakan seluruh kata (bukan cuma vocab). Karena Doc 0 dan Doc 1 memiliki kata-kata yang sangat berbeda dalam kalimat aslinya (meskipun keduanya punya kata 'sanksi'), jika fungsi split() membedakan "sanksi)," (dengan koma/kurung) dan "sanksi," maka mesin menganggap tidak ada kata yang sama, sehingga jaraknya maksimal (1.0).
- Doc 0 ke Doc 2 (0.95): Menunjukkan ada kemiripan yang sangat tipis jika menghitung seluruh kata dalam kalimat.

```python
import pandas as pd
from collections import Counter

# Data dari slide
documents = [
    "sanksi",          
    "sanksi alumni",      
    "sanksi alumni alumni"
]

# Membuat DataFrame TF secara manual
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in documents],
).fillna(0).astype(int)

# Menampilkan matriks
print("Matriks Term Frequency (TF):")
print(tf)
```

Hasil:
Tabel diatas menunjukkan perubahan proporsi, dimana:
- Doc 0 murni bicara tentang sanksi
- Doc 1 menggabung sanksi dan alumni secara seimbang
- Doc 2 lebih condong membahas mengenai alumni

```python
full_cos_sim_matrix = cosine_similarity(X)

# Display the full matrix as a DataFrame for better readability
cos_sim_df = pd.DataFrame(full_cos_sim_matrix, index=[f'Doc {i}' for i in range(len(corpus))], columns=[f'Doc {i}' for i in range(len(corpus))])
print("Full Cosine Similarity Matrix:")
print(cos_sim_df)
```
